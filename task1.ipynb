```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
penguin = sns.load_dataset('penguins')
```

```python
sns.head()
```

```python
fig = plt.figure(figsize=(10,6))
ax = sns.barplot(x=penguin["species"], y=penguin["bill_length_mm"], data=penguin)
plt.show()
```

```python
fig1 = plt.figure(figsize=(10,6))
ax1=sns.distplot(penguin["bill_depth_mm"].values,bins=20)
```

```python
fig2 = plt.figure(figsize=(10,6))
ax2 = sns.scatterplot(x=penguin["sex"], y=penguin["body_mass_g"], data=penguin)
plt.show()
```

```python
fig3 = plt.figure(figsize=(14,12))
ax3 = fig.add_subplot(221)
ax3 = sns.lineplot(x=penguin["bill_depth_mm"], y=penguin["flipper_length_mm"], data=penguin)
ax3.set_title('')
```
