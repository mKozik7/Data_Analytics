```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.datasets import make_blobs
```

```python
penguin = sns.load_dataset('penguins')
penguin
```

```python
sns.pairplot(penguin)
```

```python
penguin1 = penguin.dropna()
penguin1 = penguin1.drop(columns=["species","sex","island"])
kmeans = KMeans(n_clusters=3, random_state=0).fit_predict(penguin1)
```

```python
penguin1["kmeans"] = kmeans
```

```python
sns.pairplot(penguin1, hue="kmeans")
```

```python
sns.pairplot(penguin, hue="island")
```

```python
sns.pairplot(penguin, hue="species")
```

```python
sns.pairplot(penguin, hue="sex")
```
